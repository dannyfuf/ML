```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB
```

# Predict Upvotes

## Import Dataset

```python
data_upvotes = pd.read_csv('Datasets/upvotes/train_upvotes.csv')
data_upvotes.head(1)
```

## Preprocessing

```python
y = data_upvotes['Upvotes']
y
```

```python
data_upvotes.drop(columns = ['Upvotes'])
```

```python
categories = pd.get_dummies(data_upvotes['Tag'])
categories
```

```python
numbers = data_upvotes[['Reputation', 'Answers', 'Username', 'Views']]
columns = numbers.columns
```

```python
scaler = StandardScaler()
```

```python
numbers = scaler.fit_transform(numbers)
```

```python
numbers = pd.DataFrame(numbers, columns = columns)
numbers.head(3)
```

```python
X = pd.concat([categories, numbers], axis=1).reindex(categories.index)
X.head(3)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

## Model

```python
discriminative = LogisticRegression(random_state=0)
generative = BernoulliNB(alpha=1)
```

```python
discriminative.fit(X_train, y_train)
```

```python
generative.fit(X_train, y_train)
```
